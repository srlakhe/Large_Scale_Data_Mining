```python
import json
import os
from os import listdir
#import matplotlib.pyplot as plt
import datetime, time
import pytz
#import statsmodels.api as sm
import numpy as np
import pandas as pd
```

```python
def file_len(fname):
    with open(fname) as f:
        for i, l in enumerate(f):
            pass
    return i + 1
```

```python
tweet_data_filenames = os.listdir("D:/Git/Data_for_project5/test_data")
#tweet_data_filenames = ["a.txt"]
print(tweet_data_filenames)

pst_tz = pytz.timezone('US/Pacific')
for file_num, file in enumerate(tweet_data_filenames):
    with open("D:/Git/Data_for_project5/test_data/" + file,'r',encoding = 'utf-8') as f:
        
        print('opening done ' + file)
        for length,l in enumerate(f):
            pass
        print("number of lines = " + str(length))
    
        df = pd.DataFrame(0, index=np.arange(length), columns=['time_of_tweet', 'num_of_tweets', 'num_retweets', 'sum_followers', 'max_followers','sum_fav_count','sum_verified','sum_status','sum_friends','sum_ranking_score','sum_impressions','sum_momentum'])
        #df = pd.DataFrame(columns = ['time_of_tweet', 'num_of_tweets', 'num_retweets', 'sum_followers', 'max_followers'])
        with open("D:/Git/Data_for_project5/test_data/" + file,'r',encoding = 'utf-8') as f:
            for i,line in enumerate(f):
                if i % 5000 == 0:
                    print(str(i) + " lines completed")
                    #break
                tweet = json.loads(line)
                time_of_tweet = datetime.datetime.fromtimestamp(tweet['firstpost_date'],pst_tz)
                num_of_tweets = 1
                num_retweets = tweet['metrics']['citations']['total']
                sum_followers = tweet['author']['followers']
                max_followers = tweet['author']['followers']
                sum_fav_count = tweet['tweet']['favorite_count']
                #print()
                #print(tweet['tweet']['user']['verified'] == False)
                if tweet['tweet']['user']['verified'] == False:
                    sum_verified = 0
                else:
                    sum_verified = 1
                sum_status = tweet['tweet']['user']['statuses_count']
                sum_friends = tweet['tweet']['user']['friends_count']
                #sum_influence = tweet['original_author']['influence_level']
                sum_ranking_score = tweet['metrics']['ranking_score']
                sum_impressions = tweet['metrics']['impressions']
                sum_momentum = tweet['metrics']['momentum']

                df.set_value(i,'time_of_tweet', time_of_tweet)
                df.set_value(i,'num_of_tweets', num_of_tweets)
                df.set_value(i,'num_retweets', num_retweets)
                df.set_value(i,'sum_followers', sum_followers)
                df.set_value(i,'max_followers', max_followers)
                df.set_value(i,'sum_fav_count', sum_fav_count)
                df.set_value(i,'sum_verified', sum_verified)
                df.set_value(i,'sum_status', sum_status)
                df.set_value(i,'sum_friends', sum_friends)
                #df.set_value(i,'sum_influence', sum_influence)
                df.set_value(i,'sum_ranking_score', sum_ranking_score)
                df.set_value(i,'sum_impressions', sum_impressions)
                df.set_value(i,'sum_momentum', sum_momentum)
                
                
                
                
            df.to_csv("df_part1e_" + file[:-4] + ".csv")
```
